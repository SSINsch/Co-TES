```python
p = r'C:\Users\SHIN-DESKTOP\Documents\remote\diff_model_coteaching_plus\logs\debug_20221015.log'
with open(p, "r") as f:
    lines = f.readlines()
lines = ' '.join(lines)
lines = lines.split('[Args]')
tests = lines[1:]
```

```python
results = []
for test in tests:
    logs = test.split('\n')
    raw_results = []
    title = []
    for lg in logs:
        if 'Test results' in lg:
            raw_results.append(lg)
        elif 'seed' in lg:
            title.append(lg)

    title = title[0][1:]
    temp = {title: []}

    for r in raw_results:
        acc_1 = r[r.find('Acc_1: ')+7:r.find('Acc_1: ')+13]
        acc_2 = r[r.find('Acc_2: ')+7:r.find('Acc_2: ')+13]
        loss_1 = r[r.find('Loss_1: ')+8:r.find('Loss_1: ')+14]
        loss_2 = r[r.find('Loss_2: ')+8:r.find('Loss_2: ')+14]
        temp[title].append({'Acc_1': acc_1, 'Acc_2': acc_2, 'Loss_1': loss_1, 'Loss_2': loss_2})

    results.append(temp)
```

```python
for r_k in results:
    print(list(r_k.keys())[0])
    for k in r_k.keys():
        for i in r_k[k]:
            print(f'\t{i}')
```
